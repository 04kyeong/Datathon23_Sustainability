### Light Charts

```python
# rename the columns for light
Light1 = df["I turn off the light when I'm out of the room."]

# pie chart
Light1.value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Percent of turning off the light (2023)')
plt.ylabel('Turning off light when living the room')
```

```python
# rename the columns for light2
Light2 = df["I turn off the light when I'm out of the room..1"]

# pie chart
Light2.value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Percent of turning off the light (last 5 years)')
plt.ylabel('Turning off light when living the room')
```

```python
# rename the columns for light3
Light3 = df["I turn off the light when I'm out of the room..2"]

# pie chart
Light3.value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Percent of turning off the light (last 10 years)')
plt.ylabel('Turning off light when living the room')
```

```python
L1 = Light1.value_counts()
L2 = Light2.value_counts()
L3 = Light3.value_counts()

Yes = [L1[0],L2[0],L3[0]]
No = [L1[1],L2[1],L3[1]]
Sometimes = [L1[2],L2[2],L3[2]]
index = ['Light1', 'Light2', 'Light3']
df = pd.DataFrame({'Yes': Yes,
                  'No': No,
                      'Sometimes':Sometimes}, index=index)
ax = df.plot.bar(rot=0)
```

```python
# download the word cloud
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
#clean data
df['Motivation'] =(df['If you selected others on previous question, please write what it was.'].astype(str))
#df['Motivation'] = df['Motivation'].replace('', np.nan)

#df['If you selected others on previous question, please write what it was.'] = df['If you selected others on previous question, please write what it was.'].dropna()
df = df.dropna()
print(df['Motivation'])
```

```python
# text data
text = ' '.join(df['Motivation'].astype(str))
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display the word cloud using matplotlib
plt.figure(figsize=(15, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud for suggestion of the method to increase people's participation in sustainability challenge")
plt.show()
```
